```python
import matplotlib.pyplot as plt
import csv
```

```python
X = []
Y = []

plt.figure(figsize=(15, 9))

Y = [4, 8, 16, 32, 64, 128, 256, 512]
X = []
with open('TTAS.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
     
    for ROWS in plotting:
        X.append(float(ROWS[0]))
plt.scatter(Y, X, label = 'TTAS', s=100, alpha=0.7)

X = []
with open('TAS.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
     
    for ROWS in plotting:
        X.append(float(ROWS[0]))
plt.scatter(Y, X, label = 'TAS', s=100, alpha=0.7)
X = []

Y = [4, 8, 16, 32, 64, 128, 256]
with open('ticket.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
     
    for ROWS in plotting:
        X.append(float(ROWS[0]))
plt.scatter(Y, X, label = 'ticket', s=100, alpha=0.7)

plt.title('Compare average time per operation')
plt.xlabel('Size')
plt.ylabel('time(sec)')
plt.ylim([-0.00001, 0.0003])
plt.legend()
plt.grid()
plt.show()
```
